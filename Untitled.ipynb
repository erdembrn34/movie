```python
import pandas as pd

df = pd.read_csv("Top_10000_Movies_IMDb.csv")
print(df.head())
print(df.info())
```

```python
df = df.dropna(subset=["Rating"])
df["Rating"] = df["Rating"].astype(float)
```

```python
print(df.columns)
```

```python
# En yüksek puanlı 10 filmi seç
top10 = df.sort_values("Rating", ascending=False).head(10)
print(top10[["Movie Name", "Rating"]])

# Görselleştirme
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(top10["Movie Name"], top10["Rating"], color="skyblue")
plt.xlabel("IMDb Rating")
plt.ylabel("Movie Name")
plt.title("Top 10 Movies by Rating")
plt.gca().invert_yaxis()  # en yüksek puan üstte
plt.show()
```

```python
# Türlere göre ortalama puan
genre_ratings = df.groupby("Genre")["Rating"].mean().sort_values(ascending=False)

print(genre_ratings.head(10))  # en yüksek puanlı 10 tür

# Görselleştirme
plt.figure(figsize=(12,6))
genre_ratings.head(10).plot(kind="bar", color="orange")
plt.ylabel("Average Rating")
plt.title("Top 10 Genres by Average IMDb Rating")
plt.show()
```

```python
# En çok oylanan 10 film
top10_votes = df.sort_values("Votes", ascending=False).head(10)
print(top10_votes[["Movie Name", "Votes"]])

plt.figure(figsize=(10,6))
plt.barh(top10_votes["Movie Name"], top10_votes["Votes"], color="green")
plt.xlabel("Number of Votes")
plt.ylabel("Movie Name")
plt.title("Top 10 Most Voted Movies")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Metascore boş değerleri temizle
df_clean = df.dropna(subset=["Metascore"])

plt.figure(figsize=(8,6))
plt.scatter(df_clean["Rating"], df_clean["Metascore"], alpha=0.5)
plt.xlabel("IMDb Rating")
plt.ylabel("Metascore")
plt.title("IMDb Rating vs Metascore")
plt.show()
```
